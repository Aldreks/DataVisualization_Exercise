```python
import matplotlib.pyplot as pp
```

```python
meses = ["jan", "fev", "mar", "abr", "mai", "jun", "jul", "ago", "set", "out", "nov", "dez"]
fase1 = [850, 320, 980, 560, 430, 710, 275, 370, 900, 880, 950, 930]
fase2 = [600, 450, 680, 250, 230, 520, 220,  540, 735, 666, 700, 500]

pp.figure(figsize=(12, 6))
pp.style.use("dark_background")   #('ggplot')
pp.grid(False)   #Impoe grade no grafico

pp.suptitle(f"PYTHON 3.9.5  --   Aldreks Albuquerque", size=11)
pp.title(f"GR√ÅFICO DE LINHAS  -  LEITURAS POR FASE\n", fontweight="bold", size = 13)
pp.xlabel("Meses",   color="yellow", fontweight="bold", size = 14)
pp.ylabel("Leituras", color="yellow", fontweight="bold", size = 14)

pp.plot(meses, fase1,  color="red",  label="Fase1", marker="*", linewidth=1)
pp.plot(meses, fase1,  color="white",  linewidth=8, alpha=0.1)

pp.plot(meses, fase2,  color="yellow", label="Fase2", marker="o", linewidth=2)
pp.plot(meses, fase2,  color="blue", linewidth=8, alpha=0.3)

# Cores dos Eixos X/Y
pp.tick_params(axis="x", colors="yellow")
pp.tick_params(axis="y", colors="yellow")

pp.legend()
pp.show()
```
